```
#task2  how many of the men and women have blood glucsoe level more than 200,
#have smoking habit and smoking history,have hypertension,have heart disease and
# he/she should not have diabetese
df[(df["blood_glucose_level"]>=200) & (df["smoking_history"].isin(["current","former","ever","not current"])) & (df["hypertension"]==1) & (df["heart_disease"]==1) &(df["diabetes"]==0)].value_counts("gender")
```

```
#task3  along with task 2 add age below 14 and perform how many male and female are in diabetes
df[(df["blood_glucose_level"]>=200) & (df["smoking_history"].isin(["current","former","ever","not current"])) &
(df["hypertension"]==1) & (df["heart_disease"]==1) &(df["diabetes"]==1) & (df["age"]<=14)].value_counts("gender")
```

```
#task4  how many mele and female have perfect bmi, glucose < 180 ,no smoking habit,no bp,no hd and age > 40
df[(df["bmi"].between(18.5,24.9))&(df["blood_glucose_level"] < 180) & (df["smoking_history"]=="never")& (df["hypertension"]==0)&(df["heart_disease"]==0) & (df["age"]>40)].value_counts("gender")
```

```
#task5 lean or obese and hbaic<6 from all these kids(<12), and teenager(12-18) and adult(18)

def age_group(age):
  if age < 13:
    return "Child"
  elif age > 12 and age < 20:
    return "Teenager"
  else:
    return "Elder"


df["age_catagory"] = df["age"].apply(age_group)
```

```
df.head()
```

```
def bmi_category(bmi):
  if bmi < 18.5:
    return "Underweight"
  elif 18.5 <= bmi < 25:
    return "Normal"
  elif 25 <= bmi < 30:
    return "Overweight"
  else:
    return "Obese"

df['bmi_category'] = df['bmi'].apply(bmi_category)
```

```
df.drop("age_catagory",axis=1,inplace=True)
```
